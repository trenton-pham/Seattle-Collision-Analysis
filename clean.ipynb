```python
import geopandas as gpd
import pandas as pd
import datetime as dt
```

### All Collisions Dataset
This dataset contains all types of collisions from pedestrians, cyclists, and vehicles involved with specific latitude and longtitude coordinates that will be used for our geospatial analysis later.

Note that not all columns will be used in our EDA.

```python
gdf = gpd.read_file('data/raw/SDOT_Collision_All_Years.geojson')

gdf['INCDATE'] = pd.to_datetime(gdf['INCDATE'])
gdf['YEAR'] = gdf['INCDATE'].dt.year
gdf = gdf[(gdf["YEAR"] >= 2015) & (gdf['YEAR'] <= 2025)]

cols_to_drop = [
    'EXCEPTRSNCODE',    
    'EXCEPTRSNDESC',    
    'INATTENTIONIND',   
    'PEDROWNOTGRNT',  
    'SPEEDING',          
    'SHAREDMICROMOBILITYCD',  
    'SHAREDMICROMOBILITYDESC',
    'SDOTCOLNUM',    
    'SPDCASENO',  
    'ST_COLCODE', 
    'ST_COLDESC',
    'REPORTNO',
    'SE_ANNO_CAD_DATA',
    'INCKEY',
    'STATUS',
    'ADDRTYPE',
    'DIAGRAMLINK',
    'SOURCEDESC',
    'ADDDTTM',
    'MODDTTM',
    'OBJECTID',
    'INTKEY',
    'COLLISIONTYPE',
    'REPORTLINK',
    'SEGLANEKEY',
    'CROSSWALKKEY',
    'HITPARKEDCAR',
    'SOURCE'
]

gdf.drop(columns=cols_to_drop, inplace=True)

moderate_nan_cols = ['UNDERINFL', 'WEATHER', 'ROADCOND', 'LIGHTCOND', 'JUNCTIONTYPE']

for col in moderate_nan_cols:
    gdf[col].fillna('Unknown', inplace=True)

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
gdf['Day_of_Week'] = gdf['INCDATE'].dt.day_name()

def get_season(month):
    if ((month > 0) & (month < 3)) | (month == 12):
        return 'Winter'
    elif (month > 2) & (month < 6):
        return 'Spring'
    elif (month > 5) & (month < 9):
        return 'Summer'
    elif (month > 8) & (month < 12):
        return 'Fall'

gdf = gdf.assign(Season=gdf.get('INCDATE').dt.month.apply(get_season))
gdf

gdf.to_file('data/cleaned/Collision_All_Filtered.geojson', driver='GeoJSON')
```
